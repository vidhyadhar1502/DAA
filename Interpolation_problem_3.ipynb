```python
import time
import random

def interpolation_search(arr, key):
    
    low = 0
    high = len(arr) - 1
    comparison = 0

    while low <= high and key >= arr[low] and key <= arr[high]:
        comparison += 1
        
        if arr[low] == arr[high]:
            if arr[low] == key:
                return low, probes
            break

        # Interpolation formula for estimating position
        pos = low + int(((float(high - low) / (arr[high] - arr[low])) * (key - arr[low])))

        # Out of bounds check (can happen with extreme non-uniform distributions)
        if pos < low or pos > high:
            break

        if arr[pos] == key:
            return pos, comparison
        
        if arr[pos] < key:
            low = pos + 1
        else:
            high = pos - 1

    return -1, comparison

def generate_dataset():
    print("\n--- Dataset Configuration ---")
    print("1. Uniformly Distributed")
    print("2. Non-Uniformly Distributed (Exponential-like)")
    dist_choice = input("Select data distribution (1 or 2): ").strip()

    print("\n1. Enter elements manually")
    print("2. Generate elements randomly")
    gen_choice = input("Select generation method (1 or 2): ").strip()

    arr = []
    
    if gen_choice == '1':
        # Manual entry
        elements_str = input("Enter numbers separated by spaces: ")
        try:
            arr = [float(x) for x in elements_str.split()]
        except ValueError:
            print("Invalid input. Please enter numbers only.")
            return None, None
    else:
        # Random generation
        try:
            size = int(input("Enter the number of elements to generate: "))
            if size <= 0:
                print("Size must be greater than zero.")
                return None, None
        except ValueError:
            print("Invalid size.")
            return None, None

        if dist_choice == '1':
            # Uniform
            start = random.randint(1, 100)
            step = random.randint(2, 10)
            arr = [start + (i * step) for i in range(size)]
        else:
            # Non-uniform
            arr = [float(i**2 + random.randint(1, 10)) for i in range(1, size + 1)]

    arr.sort()
    
    dist_type = "Uniform" if dist_choice == '1' else "Non-Uniform"
    return arr, dist_type


print("=== Interpolation Search Performance Analysis ===")
    
arr, dist_type = generate_dataset()

print(f"\nGenerated Sorted Dataset ({dist_type}):")
print(arr)

key = float(input("\nEnter the search key: "))

# Measure execution time
start_time = time.perf_counter()
index, comparison = interpolation_search(arr, key)
end_time = time.perf_counter()
    
execution_time_ms = (end_time - start_time) * 1000

print("\n--- Search Results ---")
if index != -1:
    print(f"Result: Success! Key found.")
    print(f"Index Position: {index}")
else:
    print("Result: Key NOT found.")
        
print(f"Number of Probes: {comparison}")
print(f"Execution Time: {execution_time_ms:.6f} ms")

# analysis 
print("\n--- Performance Analysis ---")
if dist_type == "Uniform":
    print("Analysis: For Uniform datasets, Interpolation Search behaves optimally.")
    print("The formula predicts positions accurately, yielding near O(log(log n)) comparison complexity.")
else:
    print("Analysis: For Non-Uniform datasets, the formula's accuracy drops.")
    print("Large gaps warp the ratio logic, causing higher probes. Worst-case can degrade toward O(n).")
```
